Next, we'll import the necessary libraries and load the pre-trained GPT-2 model and its tokenizer. For fine-tuning, you'll need a custom dataset. For demonstration purposes, I will show how to load a generic text file. You will need to replace `your_custom_dataset.txt` with the path to your actual dataset file.

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from datasets import Dataset

# Load pre-trained GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# GPT-2 does not have a default padding token, so we add one
tokenizer.pad_token = tokenizer.eos_token
```

Now, let's prepare the dataset. I'll include a placeholder for creating a dummy dataset if you don't have one ready, and then show how to load a text file and process it.

```python
# Create a dummy text file for demonstration if you don't have one
dummy_text_data = """
Once upon a time, in a land far, far away, lived a brave knight. The knight embarked on a perilous journey to save the princess. Dragons, goblins, and evil wizards tried to stop him, but his courage knew no bounds. He fought valiantly and, in the end, rescued the princess and they lived happily ever after.

Another story begins now. The sun rose over the enchanted forest. A young wizard apprentice was practicing spells. He accidentally turned his master's cat into a teacup. Oh dear, he thought, this is going to be a long day. He quickly tried to reverse the spell, but only managed to make the teacup float and sing.
"""

with open('your_custom_dataset.txt', 'w') as f:
    f.write(dummy_text_data)

# Load your custom dataset using the datasets library
# This creates a dataset from the text file, where each line is an example.
with open('your_custom_dataset.txt', 'r') as f:
    lines = [line.strip() for line in f if line.strip()]

raw_datasets = Dataset.from_dict({'text': lines})

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True, max_length=128)

tokenized_datasets = raw_datasets.map(
    tokenize_function, batched=True, num_proc=1, remove_columns=raw_datasets.column_names
)

# Combine into blocks for language modeling
def group_texts(examples):
    # Concatenate all texts.
    concatenated_examples = {k: sum(examples[k], []) for k in examples.keys()}
    total_length = len(concatenated_examples[list(examples.keys())[0]])
    # We drop the small remainder, we could add padding if the model supported it instead of this drop, you can customize this part to your needs.
    total_length = (total_length // 128) * 128
    # Split by chunks of max_len.
    result = {
        k: [t[i : i + 128] for i in range(0, total_length, 128)]
        for k, t in concatenated_examples.items()
    }
    result["labels"] = result["input_ids"].copy()
    return result


train_dataset = tokenized_datasets.map(
    group_texts,
    batched=True,
    batch_size=1000,
    num_proc=1,
)

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False
)
```

Next, we'll define the training arguments and fine-tune the GPT-2 model using the `Trainer` API. You might need to adjust parameters like `per_device_train_batch_size`, `num_train_epochs`, and `learning_rate` based on your dataset size and computational resources.

```python
training_args = TrainingArguments(
    output_dir='gpt2-finetuned',
    num_train_epochs=3,
    per_device_train_batch_size=4, # Adjust based on your GPU
    save_steps=10_000,
    save_total_limit=2,
    prediction_loss_only=True,
    logging_dir='./logs',
    logging_steps=500,
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
)

trainer.train()

# Save the fine-tuned model and tokenizer
model.save_pretrained('gpt2-finetuned')
tokenizer.save_pretrained('gpt2-finetuned')
```

Finally, let's see how to generate text using your newly fine-tuned GPT-2 model. You can adjust the `max_length` and `num_return_sequences` parameters to control the output.

```python
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer

# Load the fine-tuned model and tokenizer from the local directory
finetuned_tokenizer = GPT2Tokenizer.from_pretrained('gpt2-finetuned', local_files_only=True)
finetuned_model = GPT2LMHeadModel.from_pretrained('gpt2-finetuned', local_files_only=True)

# Create a text generation pipeline
gen = pipeline('text-generation', model=finetuned_model, tokenizer=finetuned_tokenizer)

# Generate text
prompt = "Once upon a time,"
output = gen(prompt, max_length=100, num_return_sequences=1, do_sample=True, top_k=50, top_p=0.95)

for i, sample_text in enumerate(output):
    print(f"Generated Text {i+1}:\n{sample_text['generated_text']}\n---")
```
